## 1. ¿Qué son las Expresiones Regulares?

Las **expresiones regulares** (regex o regexp) son secuencias de caracteres que definen un **patrón de búsqueda**. Son una herramienta poderosa para:

- Validar formatos de datos (emails, teléfonos, códigos postales)
- Buscar y extraer información de textos
- Reemplazar patrones en cadenas
- Limpiar y normalizar datos

```
┌─────────────────────────────────────────────────────────────────────┐
│                    EXPRESIONES REGULARES                            │
├─────────────────────────────────────────────────────────────────────┤
│                                                                     │
│   TEXTO DE ENTRADA          PATRÓN REGEX         RESULTADO          │
│   ─────────────────         ────────────         ─────────          │
│                                                                     │
│   "Mi correo es             \w+@\w+\.\w+         juan@mail.com      │
│    juan@mail.com"                                  ✓ MATCH          │
│                                                                     │
│   "Código: ABC-123"         [A-Z]{3}-\d{3}       ABC-123            │
│                                                    ✓ MATCH          │
│                                                                     │
│   "Tel: 55-1234-5678"       \d{2}-\d{4}-\d{4}   55-1234-5678       │
│                                                    ✓ MATCH          │
│                                                                     │
└─────────────────────────────────────────────────────────────────────┘
```

### ¿Por qué son importantes en Ciencia de Datos?

| Aplicación | Ejemplo |
|------------|----------|
| Limpieza de datos | Extraer números de textos mixtos |
| Validación | Verificar formato de emails en un dataset |
| Extracción | Obtener fechas de documentos no estructurados |
| Normalización | Estandarizar formatos de teléfonos |

## 2. El Módulo `re` de Python

Python incluye el módulo `re` para trabajar con expresiones regulares.

```
┌────────────────────────────────────────────────────────────────────┐
│                    FUNCIONES PRINCIPALES DE re                     │
├────────────────────────────────────────────────────────────────────┤
│                                                                    │
│   re.match()    → Busca patrón al INICIO de la cadena              │
│                   Retorna: Match object o None                     │
│                                                                    │
│   re.search()   → Busca patrón en CUALQUIER parte                  │
│                   Retorna: Match object o None                     │
│                                                                    │
│   re.findall()  → Encuentra TODAS las coincidencias                │
│                   Retorna: Lista de strings                        │
│                                                                    │
│   re.sub()      → Busca y REEMPLAZA coincidencias                  │
│                   Retorna: String modificado                       │
│                                                                    │
│   re.split()    → Divide cadena usando patrón como separador       │
│                   Retorna: Lista de strings                        │
│                                                                    │
└────────────────────────────────────────────────────────────────────┘
```

````python
import re

# Texto de ejemplo
texto = "Mi número es 55-1234-5678 y mi código es ABC-123. Otro telefono 55-1357-2468."

# re.search() - Busca en cualquier parte
resultado = re.search(r'\d{2}-\d{4}-\d{4}', texto)
print(resultado)
if resultado:
    print(f"Teléfono encontrado: {resultado.group()}")

# re.findall() - Encuentra todos los números
numeros = re.findall(r'\d+', texto)
print(f"Todos los números: {numeros}")
````

### Raw Strings: El Prefijo `r`

En Python, usamos el prefijo `r` antes de las expresiones regulares:

```python
# Con r (raw string) - RECOMENDADO
patron = r'\d+'

# Sin r - Necesitas escapar las barras
patron = '\\d+'
```

El prefijo `r` evita que Python interprete `\` como caracteres de escape.

````python
# Diferencia entre raw string y string normal
print("Sin raw string: \\d+ se imprime como:", '\d+')
print("Con raw string: r'\\d+' se imprime como:", r'\d+')

# En regex siempre usa raw strings
texto = "Hay 42 respuestas"
patron = r'\d+'
resultado = re.findall(patron, texto)
print(f"\nBúsqueda con {patron!r}: {resultado}")
````

## 3. Metacaracteres Básicos

Los metacaracteres son caracteres especiales con significado especial en regex.

```
┌─────────────────────────────────────────────────────────────────────┐
│                      METACARACTERES BÁSICOS                         │
├──────────┬──────────────────────────────────────────────────────────┤
│ Carácter │ Significado                                              │
├──────────┼──────────────────────────────────────────────────────────┤
│    .     │ Cualquier carácter (excepto nueva línea)                 │
│    ^     │ Inicio de la cadena                                      │
│    $     │ Fin de la cadena                                         │
│    \d    │ Dígito (equivale a [0-9])                                │
│    \D    │ NO dígito (equivale a [^0-9])                            │
│    \w    │ Carácter de palabra [a-zA-Z0-9_]                         │
│    \W    │ NO carácter de palabra                                   │
│    \s    │ Espacio en blanco (espacio, tab, newline)                │
│    \S    │ NO espacio en blanco                                     │
│    \b    │ Límite de palabra (word boundary)                        │
│    \\    │ Barra invertida literal                                  │
└──────────┴──────────────────────────────────────────────────────────┘
```

````python
import re

texto = "Python 3.11 fue lanzado en 2022. 3A11"

# \d - Encuentra dígitos
digitos = re.findall(r'\d', texto)
print(f"Dígitos individuales: {digitos}")

# \d+ - Encuentra secuencias de dígitos
numeros = re.findall(r'\d+', texto)
print(f"Números completos: {numeros}")

# \w+ - Encuentra palabras
palabras = re.findall(r'\w+', texto)
print(f"Palabras: {palabras}")

# . - Cualquier carácter
patron = re.findall(r'3.11', texto)
print(f"Patrón 3.11: {patron}")
````

````python
# ^ y $ - Anclas de inicio y fin
lineas = ["python es genialrrrrrrrrrrrrrrr", "Java es popularrrrrrrrrrrrrrr", "Python rocks", "Cython is cool"]

print("Líneas que EMPIEZAN con 'Python':")
for linea in lineas:
    if re.match(r'^[A-Za-z]ython', linea):
        print(f"  ✓ {linea}")

print("\nLíneas que TERMINAN con 'r':")
for linea in lineas:
    if re.search(r'r+$', linea):
        print(f"  ✓ {linea}")
````

````python
# \b - Límite de palabra (muy útil para búsquedas exactas)
texto = "El gato está en el catálogo de gatos"

# Sin \b - Encuentra 'gato' en cualquier parte

resultado1 = re.findall(r'gato', texto.lower())
print(f"Sin límite: {resultado1}")

# Con \b - Solo palabras completas
resultado2 = re.findall(r'\bgato\b', texto.lower())
print(f"Con límite de palabra: {resultado2}")

# Ejemplo práctico
codigo = "var x = 10; variable = 20;"
patron = r'\bvar\b'
resultado3 = re.findall(patron, codigo)
print(f"\nEncontrar 'var' como palabra: {resultado3}")
````

## 4. Cuantificadores

Los cuantificadores especifican **cuántas veces** debe aparecer un patrón.

```
┌─────────────────────────────────────────────────────────────────────┐
│                         CUANTIFICADORES                             │
├──────────┬──────────────────────────────────────────────────────────┤
│ Símbolo  │ Significado                                              │
├──────────┼──────────────────────────────────────────────────────────┤
│    *     │ Cero o más repeticiones                                  │
│    +     │ Una o más repeticiones                                   │
│    ?     │ Cero o una repetición (opcional)                         │
│   {n}    │ Exactamente n repeticiones                               │
│   {n,}   │ Al menos n repeticiones                                  │
│   {n,m}  │ Entre n y m repeticiones                                 │
└──────────┴──────────────────────────────────────────────────────────┘

┌─────────────────────────────────────────────────────────────────────┐
│                     EJEMPLOS VISUALES                               │
├─────────────────────────────────────────────────────────────────────┤
│                                                                     │
│   Patrón: \d+       Texto: "abc123def456"                           │
│   Resultado: ['123', '456']                                         │
│                                                                     │
│   Patrón: \d{3}     Texto: "abc123def456789"                        │
│   Resultado: ['123', '456', '789']                                  │
│                                                                     │
│   Patrón: \d{2,4}   Texto: "1 12 123 1234 12345"                    │
│   Resultado: ['12', '123', '1234', '1234']                          │
│                                                                     │
└─────────────────────────────────────────────────────────────────────┘
```

````python
import re

# * - Cero o más
print("Cuantificador * (cero o más):")
texto = "a c ab  bc cba ac abc abbc abbbc abbbbbbbbbbbbbbbbbcvv"
resultado = re.findall(r'ab*c', texto)
print(f"  Patrón 'ab*c' en '{texto}': {resultado}")
````

````python
# + - Una o más
print("\nCuantificador + (una o más):")
resultado = re.findall(r'ab+c', texto)
print(f"  Patrón 'ab+c' en '{texto}': {resultado}")
````

````python

# ? - Cero o una (opcional)
print("\nCuantificador ? (cero o una):")
texto = "color colour colouur colouuuuuuur"
resultado = re.findall(r'colou?r', texto)
print(f"  Patrón 'colou?r' en '{texto}': {resultado}")
````

````python
# Cuantificadores con números específicos
texto = "1 12 123 1234 12345 123456"

print("Cuantificadores específicos:")

# Exactamente 2 dígitos
resultado_2 = re.findall(r'\b\d{2}\b', texto)
print(f"  \\d{{2}} (exactamente 2): {resultado_2}")

resultado_2 = re.findall(r'\d{2}', texto)
print(f"  \\d{{2}} (exactamente 2): {resultado_2}")
````

````python
# Cuantificadores con números específicos
texto = "1 12 123 1234 12345 123456"

# Exactamente 3 dígitos
resultado_3 = re.findall(r'\b\d{3}\b', texto)
print(f"  \\d{{3}} (exactamente 3): {resultado_3}")
````

````python
# Entre 2 y 4 dígitos
resultado_2_4 = re.findall(r'\b\d{2,4}\b', texto)
print(f"  \\d{{2,4}} (entre 2 y 4): {resultado_2_4}")
````

````python
# Cuantificadores con números específicos
texto = "1 12 123 1234 12345 123456"

# 3 o más dígitos
resultado_3_mas = re.findall(r'\b\d{3,}\b', texto)
print(f"  \\d{{3,}} (3 o más): {resultado_3_mas}")

resultado_3_mas = re.findall(r'\d{3,}', texto)
print(f"  \\d{{3,}} (3 o más): {resultado_3_mas}")
````

````python
# Ejemplo práctico: Validar formato de teléfono mexicano
telefonos = [
    "55-1234-5678",
    "551234-5678",
    "55-12345678",
    "5512345678",
    "123-456-7890"
]

# Patrón: 2 dígitos - 4 dígitos - 4 dígitos
patron = r'^\d{2}-\d{4}-\d{4}$'

print("Validación de teléfonos (formato: XX-XXXX-XXXX):")
for tel in telefonos:
    if re.match(patron, tel):
        print(f"  ✓ {tel} - Válido")
    else:
        print(f"  ✗ {tel} - Inválido")
````

## 5. Clases de Caracteres

Las clases de caracteres permiten definir **conjuntos de caracteres** válidos.

```
┌─────────────────────────────────────────────────────────────────────┐
│                     CLASES DE CARACTERES                            │
├──────────────┬──────────────────────────────────────────────────────┤
│ Sintaxis     │ Significado                                          │
├──────────────┼──────────────────────────────────────────────────────┤
│ [abc]        │ Cualquiera de los caracteres a, b, o c               │
│ [a-z]        │ Cualquier letra minúscula de a a z                   │
│ [A-Z]        │ Cualquier letra mayúscula de A a Z                   │
│ [0-9]        │ Cualquier dígito (igual que \d)                      │
│ [a-zA-Z]     │ Cualquier letra (mayúscula o minúscula)              │
│ [a-zA-Z0-9]  │ Cualquier carácter alfanumérico                      │
│ [^abc]       │ Cualquier carácter EXCEPTO a, b, o c (negación)      │
│ [^0-9]       │ Cualquier carácter que NO sea dígito                 │
└──────────────┴──────────────────────────────────────────────────────┘
```

### Nota sobre caracteres especiales dentro de []

Dentro de `[]`, la mayoría de metacaracteres pierden su significado especial:
- `[.]` coincide con un punto literal
- `[*]` coincide con un asterisco literal
- El guion `-` es especial solo cuando está entre caracteres: `[a-z]`

````python
import re

# Clases de caracteres básicas
texto = "Hola Mundo 2024!"

print("Clases de caracteres:")
print(f"  [a-z] (minúsculas): {re.findall(r'[a-z]', texto)}")
print(f"  [A-Z] (mayúsculas): {re.findall(r'[A-Z]', texto)}")
print(f"  [0-9] (dígitos): {re.findall(r'[0-9]', texto)}")
print(f"  [a-zA-Z]+ (palabras): {re.findall(r'[a-zA-Z]+', texto)}")
````

````python
# Negación con ^
texto = "ABC-123-xyz"

print("Negación con ^:")
print(f"  Texto: '{texto}'")
print(f"  [^a-z] (NO minúsculas): {re.findall(r'[^a-z]', texto)}")
print(f"  [^0-9] (NO dígitos): {re.findall(r'[^0-9]', texto)}")
print(f"  [^A-Z-] (NO mayúsculas ni guiones): {re.findall(r'[^A-Z-]', texto)}")
print(f"  [^a-z] (NO minúsculas): {re.findall(r'[^a-z]+', texto)}")
````

````python
# Ejemplo práctico: Extraer diferentes tipos de códigos
texto = "Códigos: ABC-123, XY-45, ABCD-6789, Z-1 qwe-123 AdKLDG-2345"

# Patrón: 2-4 letras mayúsculas, guion, 1-4 dígitos
patron = r'[A-Z]{2,4}-[0-9]{1,4}'

codigos = re.findall(patron, texto)
print(f"Códigos encontrados: {codigos}")
````

````python
# Clase de caracteres personalizada para vocales
texto = "Programación para Ciencia de Datos"

# Encuentra vocales (con y sin acento)
vocales = re.findall(r'[aeiouáéíóú]', texto.lower())
consonantes = re.findall(r'[bcdfghjklmnñpqrstvwxyz]', texto.lower())

print(f"Vocales ({len(vocales)}): {vocales}")
print(f"Consonantes ({len(consonantes)}): {consonantes}")
````

## 6. Funciones del Módulo `re` en Detalle

### 6.1 `re.match()` - Buscar al inicio

````python
import re

# re.match() solo busca al INICIO de la cadena
texto = "Python es genial"

# Funciona - 'Python' está al inicio
resultado = re.match(r'Python', texto)
print(f"match('Python'): {resultado}")

# No funciona - 'genial' no está al inicio
resultado = re.match(r'genial', texto)
print(f"match('genial'): {resultado}")

# Para buscar en cualquier parte, usa search()
resultado = re.search(r'genial', texto)
print(f"search('genial'): {resultado}")
````

### 6.2 `re.search()` - Buscar en cualquier parte

````python
# re.search() busca la primera coincidencia en cualquier parte
texto = "El precio es $150.0.0 USD y $200.00 MXN"

# Encuentra el primer precio
resultado = re.search(r'\$[\d.]+', texto)
print(resultado)
if resultado:
    print(f"Primera coincidencia: {resultado.group()}")
    print(f"Posición inicio: {resultado.start()}")
    print(f"Posición fin: {resultado.end()}")
    print(f"Span (inicio, fin): {resultado.span()}")
````

### 6.3 `re.findall()` - Encontrar todas las coincidencias

````python
# re.findall() retorna una lista con TODAS las coincidencias
texto = "Contactos: juan@email.com, maria@correo.mx, pedro@mail.org"

# Patrón simple para emails
emails = re.findall(r'\w+@\w+\.\w+', texto)
print(f"Emails encontrados: {emails}")

# Encontrar todos los dominios
dominios = re.findall(r'@(\w+\.\w+)', texto)
print(f"Dominios: {dominios}")
````

### 6.4 `re.sub()` - Buscar y reemplazar

````python
# re.sub() reemplaza coincidencias con nuevo texto
texto = "Mi teléfono es 55-1234-5678 y mi otro número es 33-8765-4321"

# Ocultar números de teléfono
censurado = re.sub(r'\d{2}-\d{4}-\d{4}', 'XX-XXXX-XXXX', texto)
print(f"Original: {texto}")
print(f"Censurado: {censurado}")

# Normalizar espacios múltiples
texto_espacios = "Hola    mundo   Python    rocks"
normalizado = re.sub(r'\s+', ' ', texto_espacios)
print(f"\nCon espacios extra: '{texto_espacios}'")
print(f"Normalizado: '{normalizado}'")
````

### 6.5 `re.split()` - Dividir usando patrón

````python
# re.split() divide una cadena usando un patrón como separador
texto = "manzana,naranja;plátano uva,pera"

# Dividir por múltiples separadores (coma, punto y coma, espacio)
frutas = re.split(r'[,;\s]+', texto)
print(f"Texto: '{texto}'")
print(f"Frutas: {frutas}")

# Dividir por números
texto2 = "parte1uno23dos456tres"
partes = re.split(r'\d+', texto2)
print(f"\nTexto: '{texto2}'")
print(f"Partes: {partes}")
````

## 7. Grupos de Captura

Los paréntesis `()` crean grupos que permiten extraer partes específicas.

```
┌─────────────────────────────────────────────────────────────────────┐
│                      GRUPOS DE CAPTURA                              │
├─────────────────────────────────────────────────────────────────────┤
│                                                                     │
│   Patrón: (\d{2})-(\d{4})-(\d{4})                                  │
│   Texto:  "55-1234-5678"                                            │
│                                                                     │
│   ┌────────────────────────────────────────┐                        │
│   │         55  -  1234  -  5678           │                        │
│   │         ──     ────     ────           │                        │
│   │       grupo1  grupo2   grupo3          │                        │
│   └────────────────────────────────────────┘                        │
│                                                                     │
│   match.group()  → "55-1234-5678" (coincidencia completa)          │
│   match.group(1) → "55"                                             │
│   match.group(2) → "1234"                                           │
│   match.group(3) → "5678"                                           │
│   match.groups() → ('55', '1234', '5678')                           │
│                                                                     │
└─────────────────────────────────────────────────────────────────────┘
```

````python
import re

# Extraer partes de un teléfono
texto = "Mi número es 55-1234-5678"
patron = r'(\d{2})-(\d{4})-(\d{4})'

resultado = re.search(patron, texto)

if resultado:
    print(f"Coincidencia completa: {resultado.group()}")
    print(f"Grupo 1 (área): {resultado.group(1)}")
    print(f"Grupo 2 (primera parte): {resultado.group(2)}")
    print(f"Grupo 3 (segunda parte): {resultado.group(3)}")
    print(f"Todos los grupos: {resultado.groups()}")
````

````python
# findall con grupos retorna tuplas
texto = "Fechas: 15/03/2024, 28/11/2023, 01/06/2025"
patron = r'(\d{2})/(\d{2})/(\d{4})'

fechas = re.findall(patron, texto)
print(f"Fechas encontradas: {fechas}")

for dia, mes, anio in fechas:
    print(f"  Día: {dia}, Mes: {mes}, Año: {anio}")
````

````python
# Grupos con nombre: (?P<nombre>patrón)
texto = "Email: mario@gmail.com"
patron = r'(?P<usuario>\w+)@(?P<dominio>\w+\.\w+)'

resultado = re.search(patron, texto)

if resultado:
    print(f"Usuario: {resultado.group('usuario')}")
    print(f"Dominio: {resultado.group('dominio')}")
    print(f"Como diccionario: {resultado.groupdict()}")
````

## 8. Ejemplos Prácticos para Ciencia de Datos

````python
# Ejemplo 1: Validar emails
import re

def validar_email(email):
    """Valida formato básico de email."""
    patron = r'^[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}$'
    return bool(re.match(patron, email))

emails = [
    "usuario@dominio.com",
    "nombre.apellido@empresa.mx",
    "invalido@",
    "@sinusuario.com",
    "correo@dominio",
    "mario@gmail.c",
    "rodrigo#123@YAHOO.COM"
]

print("Validación de emails:")
for email in emails:
    estado = "✓ Válido" if validar_email(email) else "✗ Inválido"
    print(f"  {email}: {estado}")
````

````python
# Ejemplo 2: Extraer información de logs
logs = """
2024-03-15 10:23:45 INFO Usuario juan123 inició sesión
2024-03-15 10:25:12 ERROR Conexión fallida para usuario maria456
2024-03-15 10:30:00 WARNING Memoria alta: 85%
2024-03-15 10:35:22 INFO Usuario pedro789 cerró sesión
"""

# Extraer fecha, nivel y mensaje
patron = r'(\d{4}-\d{2}-\d{2}) (\d{2}:\d{2}:\d{2}) (\w+) (.+)'

print("Análisis de logs:")
for match in re.finditer(patron, logs):
    fecha, hora, nivel, mensaje = match.groups()
    print(f"  [{nivel}] {fecha} {hora}: {mensaje}")
````

````python
# Ejemplo 3: Limpiar y normalizar datos
datos_sucios = [
    "  Juan   Pérez  ",
    "MARÍA    GARCÍA",
    "pedro   LÓPEZ   ",
    "  Ana   María   Ruiz  "
]

def limpiar_nombre(nombre):
    """Limpia y normaliza un nombre."""
    # Eliminar espacios al inicio y final
    nombre = nombre.strip()
    # Reemplazar múltiples espacios por uno
    nombre = re.sub(r'\s+', ' ', nombre)
    # Capitalizar correctamente
    nombre = nombre.title()
    return nombre

print("Nombres limpios:")
for nombre in datos_sucios:
    print(f"  '{nombre}' → '{limpiar_nombre(nombre)}'")
````

````python
# Ejemplo 4: Extraer datos de texto no estructurado
texto = """
Reporte de Ventas
Producto: Laptop HP - Cantidad: 5 - Precio: $15,999.00
Producto: Mouse Logitech - Cantidad: 20 - Precio: $599.50
Producto: Teclado Microsoft - Cantidad: 15 - Precio: $1,299.00
"""

patron = r'Producto: (.+?) - Cantidad: (\d+) - Precio: \$([\d,]+\.\d{2})'

print("Productos extraídos:")
print(f"{'Producto':<25} {'Cantidad':>10} {'Precio':>15}")
print("-" * 52)

for match in re.finditer(patron, texto):
    producto, cantidad, precio = match.groups()
    print(f"{producto:<25} {cantidad:>10} ${precio:>14}")
````

## 9. Resumen de Patrones Comunes

```
┌────────────────────────────────────────────────────────────────────────┐
│                    PATRONES REGEX COMUNES                              │
├────────────────────┬───────────────────────────────────────────────────┤
│ Tipo               │ Patrón                                            │
```

### Ejercicio 1: Validador de CURP
El CURP mexicano tiene el formato: `AAAA######HAAAAA##`
- 4 letras (apellidos y nombre)
- 6 dígitos (fecha de nacimiento)
- 1 letra (sexo: H/M)
- 2 letras (estado)
- 3 consonantes
- 2 caracteres (homoclave)

````python
# Tu código aquí
def validar_curp(curp):
    """Valida el formato de un CURP mexicano."""
    # Patrón: 4 letras + 6 dígitos + H/M + 2 letras + 3 consonantes + 2 caracteres
    patron = r'^[A-Z]{4}\d{6}[HM][A-Z]{2}[BCDFGHJKLMNPQRSTVWXYZ]{3}[A-Z0-9]{2}$'
    return bool(re.match(patron, curp.upper()))

# Prueba tu función
curps_prueba = [
    "GARC850101HDFRRL09",
    "PELM900515MDFRRR01",
    "ABC123",  # Inválido
]

for curp in curps_prueba:
    print(f"{curp}: {'Válido' if validar_curp(curp) else 'Inválido'}")
````

### Ejercicio 2: Extraer hashtags de un texto

### Ejercicio 3: Convertir fechas de formato MM/DD/YYYY a DD/MM/YYYY

````python
# Tu código aquí
texto = "Fechas US: 03/15/2024, 11/28/2023, 06/01/2025"

# Usa re.sub con grupos para reorganizar
texto_convertido = re.sub(r'(\d{2})/(\d{2})/(\d{4})', r'\2/\1/\3', texto)
print(f"Original: {texto}")
print(f"Convertido: {texto_convertido}")
````
