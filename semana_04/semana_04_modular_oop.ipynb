## 1. Modulos en Python

Un **modulo** es simplemente un archivo `.py` que contiene codigo Python (funciones, clases, variables).

### Por que usar Modulos?

1. **Organizacion**: Dividir codigo logicamente
2. **Reutilizacion**: Usar el mismo codigo en multiples proyectos
3. **Namespace**: Evitar conflictos de nombres
4. **Mantenibilidad**: Cambiar una parte sin afectar otras

### Crear tu Propio Modulo

Crea un archivo `operaciones.py`:

````python
# Contenido de operaciones.py
# (simulado aqui para el notebook)

operaciones_py = '''
"""Modulo con operaciones matematicas basicas."""

PI = 3.14159

def sumar(a, b):
    """Retorna la suma de dos numeros."""
    return a + b

def restar(a, b):
    """Retorna la resta de dos numeros."""
    return a - b

def area_circulo(radio):
    """Calcula el area de un circulo."""
    return PI * radio ** 2

class Calculadora:
    """Calculadora simple."""
    def __init__(self):
        self.historial = []
    
    def calcular(self, operacion, a, b):
        if operacion == "+":
            resultado = a + b
        elif operacion == "-":
            resultado = a - b
        self.historial.append(f"{a} {operacion} {b} = {resultado}")
        return resultado
'''

print(operaciones_py)
````

### Formas de Importar

````python
# FORMA 1: import modulo
import operaciones
operaciones.sumar(1, 2)  # Usar con prefijo
operaciones.PI
````

````python
# FORMA 2: import modulo as alias

import operaciones as op
op.sumar(1, 2)  # Alias mas corto
````

````python
# FORMA 3: from modulo import elemento

from operaciones import sumar, PI
resultado = sumar(4, 7)  # Sin prefijo
resultado, PI
````

````python
# FORMA 4: from modulo import *  (NO RECOMENDADO)

from operaciones import *
sumar(1, 2)  # Importa TODO - puede causar conflictos
````

````python

# Ejemplo con modulo real
import math
print(f"math.pi = {math.pi}")
print(f"math.sqrt(16) = {math.sqrt(16)}")

from math import ceil, floor
print(f"ceil(3.2) = {ceil(3.2)}")
print(f"floor(3.8) = {floor(3.8)}")
````

### El Patron `if __name__ == "__main__"`

Este patron es **fundamental** para crear modulos reutilizables.

````python
# Cuando Python ejecuta un archivo:
# - Si es el archivo principal: __name__ = "__main__"
# - Si es importado: __name__ = nombre del modulo

ejemplo_modulo = '''
# operaciones.py

def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

# Este codigo SOLO se ejecuta si corres "python operaciones.py"
# NO se ejecuta cuando haces "import operaciones"
if __name__ == "__main__":
    # Pruebas del modulo
    print("Probando operaciones...")
    print(f"sumar(2, 3) = {sumar(2, 3)}")
    print(f"restar(5, 2) = {restar(5, 2)}")
    print("Todas las pruebas pasaron!")
'''

print(ejemplo_modulo)

# Demostrar __name__ en el notebook
print(f"\n__name__ en este contexto: {__name__}")
````

## 3. Programacion Orientada a Objetos (OOP)

OOP es un paradigma de programacion que organiza el codigo en **objetos** que combinan datos y comportamiento.

### Tu Primera Clase

````python
class Producto:
    """Representa un producto en inventario."""
    
    # Atributo de clase (compartido por todas las instancias)
    iva = 0.16
    
    def __init__(self, nombre, precio, stock):
        """
        Constructor: se llama al crear un nuevo objeto.
        
        Args:
            nombre: Nombre del producto
            precio: Precio unitario
            stock: Cantidad disponible
        """
        # Atributos de instancia (unicos para cada objeto)
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
    
    def valor_inventario(self):
        """Calcula el valor total en inventario."""
        return self.precio * self.stock
    
    def precio_con_iva(self):
        """Retorna el precio con IVA incluido."""
        return self.precio * (1 + self.iva)
    
    def vender(self, cantidad):
        """Reduce el stock al vender unidades."""
        if cantidad > self.stock:
            raise ValueError(f"Stock insuficiente. Disponible: {self.stock}")
        self.stock -= cantidad
        return cantidad * self.precio

# Crear objetos (instancias)
laptop = Producto("Laptop HP", 15000, 10)
mouse = Producto("Mouse Logitech", 350, 50)

# Acceder a atributos
print(f"Producto: {laptop.nombre}")
print(f"Precio: ${laptop.precio:,}")
print(f"Stock: {laptop.stock}")

# Llamar metodos
print(f"\nValor en inventario: ${laptop.valor_inventario():,}")
print(f"Precio con IVA: ${laptop.precio_con_iva():,.2f}")

# Vender productos
venta = laptop.vender(2)
print(f"\nVenta realizada: ${venta:,}")
print(f"Stock restante: {laptop.stock}")
````

### El Parametro `self`

`self` es una referencia al objeto actual. Es como decir "yo mismo".

````python
class Contador:
    def __init__(self):
        self.valor = 0  # self.valor pertenece a ESTE objeto
    
    def incrementar(self):
        self.valor += 1  # Modifica el valor de ESTE objeto
        return self  # Permite encadenar metodos
    
    def decrementar(self):
        self.valor -= 1
        return self

# Cada objeto tiene su propio valor
contador1 = Contador()
contador2 = Contador()

contador1.incrementar()
contador1.incrementar()
contador1.incrementar()

contador2.incrementar()

print(f"Contador 1: {contador1.valor}")  # 3
print(f"Contador 2: {contador2.valor}")  # 1

# Method chaining (encadenar metodos)
contador3 = Contador()
contador3.incrementar().incrementar().incrementar().decrementar()
print(f"Contador 3: {contador3.valor}")  # 2
````

### Metodos Especiales (Dunder Methods)

````python
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
    
    def __str__(self):
        """
        Se llama cuando usas print() o str().
        Para usuarios finales: legible.
        """
        return f"{self.nombre} - ${self.precio:,} ({self.stock} en stock)"
    
    def __repr__(self):
        """
        Se llama en el REPL o al inspeccionar.
        Para desarrolladores: tecnico, reproducible.
        """
        return f"Producto('{self.nombre}', {self.precio}, {self.stock})"
    
    def __eq__(self, otro):
        """Define igualdad con ==."""
        if not isinstance(otro, Producto):
            return False
        return self.nombre == otro.nombre
    
    def __lt__(self, otro):
        """Define menor que (<). Permite ordenar."""
        return self.precio < otro.precio

# Crear productos
p1 = Producto("Laptop", 15000, 5)
p2 = Producto("Mouse", 350, 20)
p3 = Producto("Laptop", 16000, 3)  # Mismo nombre, diferente precio

# __str__: para usuarios
print(p1)

# __repr__: para desarrolladores
print(repr(p1))

# __eq__: comparacion
print(f"\np1 == p3 (mismo nombre): {p1 == p3}")
print(f"p1 == p2 (diferente nombre): {p1 == p2}")

# __lt__: permite ordenar
productos = [p1, p2, p3]
productos_ordenados = sorted(productos)  # Ordena por precio
print("\nOrdenados por precio:")
for p in productos_ordenados:
    print(f"  {p}")
````

### Atributos Privados (Convencion)

````python
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular
        self._saldo = saldo_inicial  # Convencion: _ indica "privado"
        self.__pin = "1234"  # __ hace name mangling (mas dificil de acceder)
    
    @property
    def saldo(self):
        """Getter: permite leer _saldo como cuenta.saldo"""
        return self._saldo
    
    def depositar(self, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser positiva")
        self._saldo += cantidad
        return self._saldo
    
    def retirar(self, cantidad):
        if cantidad > self._saldo:
            raise ValueError("Saldo insuficiente")
        self._saldo -= cantidad
        return self._saldo

# Uso
cuenta = CuentaBancaria("Ana Garcia", 1000)

# Leer saldo usando property
print(f"Saldo: ${cuenta.saldo:,}")  # Usa el @property

# Modificar saldo a traves de metodos (controlado)
cuenta.depositar(500)
print(f"Despues de depositar: ${cuenta.saldo:,}")

cuenta.retirar(200)
print(f"Despues de retirar: ${cuenta.saldo:,}")

# Esto funciona pero NO deberias hacerlo (rompe encapsulamiento)
# cuenta._saldo = 1000000  # Mal! Pero Python lo permite

# Esto es mas dificil de acceder (name mangling)
# cuenta.__pin  # AttributeError
# cuenta._CuentaBancaria__pin  # Funciona pero NO lo hagas
````

## 4. Herencia

Una clase puede **heredar** atributos y metodos de otra clase.

````python
# Clase base (padre)
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
    
    def valor_inventario(self):
        return self.precio * self.stock
    
    def __str__(self):
        return f"{self.nombre}: ${self.precio:,}"


# Clase derivada (hija)
class ProductoElectronico(Producto):
    """Producto con garantia."""
    
    def __init__(self, nombre, precio, stock, garantia_meses):
        # Llamar al constructor del padre
        super().__init__(nombre, precio, stock)
        # Agregar atributo propio
        self.garantia_meses = garantia_meses
    
    def __str__(self):
        # Extender el metodo del padre
        base = super().__str__()
        return f"{base} - {self.garantia_meses} meses de garantia"


class ProductoPerecible(Producto):
    """Producto con fecha de caducidad."""
    
    def __init__(self, nombre, precio, stock, fecha_caducidad):
        super().__init__(nombre, precio, stock)
        self.fecha_caducidad = fecha_caducidad
    
    def esta_vencido(self, fecha_actual):
        return fecha_actual > self.fecha_caducidad


# Uso
laptop = ProductoElectronico("Laptop HP", 15000, 5, 24)
leche = ProductoPerecible("Leche", 25, 100, "2026-03-01")

print(laptop)
print(f"Valor inventario laptop: ${laptop.valor_inventario():,}")

print(f"\n{leche}")
print(f"Esta vencida (hoy 2026-02-15): {leche.esta_vencido('2026-02-15')}")
print(f"Esta vencida (hoy 2026-04-01): {leche.esta_vencido('2026-04-01')}")
````

### Composicion vs Herencia

A veces es mejor que un objeto **contenga** otro objeto en lugar de heredar de el.

````python
# COMPOSICION: Un objeto TIENE otros objetos

class Direccion:
    def __init__(self, calle, ciudad, cp):
        self.calle = calle
        self.ciudad = ciudad
        self.cp = cp
    
    def __str__(self):
        return f"{self.calle}, {self.ciudad} {self.cp}"


class Cliente:
    def __init__(self, nombre, email):
        self.nombre = nombre
        self.email = email
        self.direcciones = []  # Cliente TIENE direcciones
    
    def agregar_direccion(self, direccion):
        self.direcciones.append(direccion)
    
    def __str__(self):
        texto = f"Cliente: {self.nombre} ({self.email})"
        for i, d in enumerate(self.direcciones, 1):
            texto += f"\n  Direccion {i}: {d}"
        return texto


# Uso
cliente = Cliente("Ana Garcia", "ana@email.com")
cliente.agregar_direccion(Direccion("Av. Principal 123", "CDMX", "06600"))
cliente.agregar_direccion(Direccion("Calle Secundaria 456", "GDL", "44100"))

print(cliente)
````

````python
# Cuando usar cada una:

guia = '''
╔════════════════════════════════════════════════════════════════════════════╗
║               HERENCIA vs COMPOSICION                                      ║
╠════════════════════════════════════════════════════════════════════════════╣
║                                                                            ║
║  HERENCIA: "Es un tipo de" (IS-A)                                          ║
║  ─────────────────────────────────                                         ║
║  - ProductoElectronico ES UN Producto                                      ║
║  - Perro ES UN Animal                                                      ║
║  - Gerente ES UN Empleado                                                  ║
║                                                                            ║
║  COMPOSICION: "Tiene un" (HAS-A)                                           ║
║  ───────────────────────────────                                           ║
║  - Cliente TIENE Direcciones                                               ║
║  - Carro TIENE Motor                                                       ║
║  - Pedido TIENE Productos                                                  ║
║                                                                            ║
║  REGLA GENERAL: Prefiere composicion sobre herencia                        ║
║  (es mas flexible y menos acoplada)                                        ║
║                                                                            ║
╚════════════════════════════════════════════════════════════════════════════╝
'''

print(guia)
````

## 5. Generadores y Procesamiento Eficiente

Para procesar archivos grandes sin cargar todo en memoria.

````python
# PROBLEMA: Leer archivo grande de golpe
# datos = open("archivo_enorme.csv").readlines()  # Carga TODO en memoria!

# SOLUCION: Generadores con yield

def leer_por_lineas(contenido):
    """Generador que procesa linea por linea."""
    for linea in contenido.split('\n'):
        # yield "pausa" la funcion y retorna el valor
        # La siguiente llamada continua desde aqui
        yield linea.strip()

# Simulacion
csv_grande = """nombre,precio,stock
Laptop,15000,10
Mouse,350,50
Teclado,800,30
Monitor,6000,15"""

# El generador NO carga todo, procesa uno a la vez
for linea in leer_por_lineas(csv_grande):
    print(f"Procesando: {linea}")
````

````python
# Procesamiento por chunks (lotes)

def procesar_en_chunks(datos, chunk_size=2):
    """Procesa datos en lotes de tamano fijo."""
    chunk = []
    for item in datos:
        chunk.append(item)
        if len(chunk) >= chunk_size:
            yield chunk  # Retorna el lote
            chunk = []   # Reinicia para el siguiente
    
    # Retornar el ultimo chunk si hay elementos
    if chunk:
        yield chunk

# Ejemplo
productos = ["A", "B", "C", "D", "E", "F", "G"]

print("Procesando en lotes de 2:")
for i, chunk in enumerate(procesar_en_chunks(productos, 2), 1):
    print(f"  Lote {i}: {chunk}")
````

````python
# Generator expressions (como list comprehension pero lazy)

# List comprehension: crea TODA la lista en memoria
cuadrados_lista = [x**2 for x in range(1000000)]
# Usa ~8MB de memoria!

# Generator expression: calcula sobre demanda
cuadrados_gen = (x**2 for x in range(1000000))
# Usa ~0 memoria hasta que iteras!

print(f"Tipo lista: {type(cuadrados_lista)}")
print(f"Tipo generador: {type(cuadrados_gen)}")

# Obtener algunos valores del generador
print(f"\nPrimeros 5 cuadrados: {[next(cuadrados_gen) for _ in range(5)]}")
````
