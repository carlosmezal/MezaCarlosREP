## Reglas de Procesamiento

### Regla 1: Conversion de Fahrenheit a Celsius

Si la unidad es `F`, convertir usando la formula:

```
        (F - 32) × 5
  C  =  ────────────
             9
```

O en Python:
```python
celsius = (fahrenheit - 32) * 5 / 9
```

### Ejemplos de Conversion

| Fahrenheit | Calculo | Celsius |
|------------|---------|----------|
| 32°F | (32-32) × 5/9 | 0°C (punto de congelacion) |
| 50°F | (50-32) × 5/9 | 10°C |
| 68°F | (68-32) × 5/9 | 20°C |
| 95°F | (95-32) × 5/9 | 35°C |
| 212°F | (212-32) × 5/9 | 100°C (punto de ebullicion) |

````python
# Funcion de conversion
def fahrenheit_a_celsius(f):
    """Convierte Fahrenheit a Celsius."""
    return (f - 32) * 5 / 9

# Pruebas de la formula
print("Tabla de conversion F -> C")
print("=" * 30)
for f in [32, 50, 68, 77, 86, 95, 104, 212]:
    c = fahrenheit_a_celsius(f)
    print(f"{f:>5}°F  =  {c:>6.1f}°C")
````

### Regla 2: Clasificacion de Temperatura

Segun la temperatura **en Celsius**, clasificar:

```
╔════════════════════════════════════════════════════════════════════════════╗
║                    TABLA DE CLASIFICACION                                  ║
╠════════════════════════════════════════════════════════════════════════════╣
║                                                                            ║
║   Temperatura (°C)          Clasificacion       Ejemplo                    ║
║   ────────────────          ─────────────       ───────                    ║
║   < 0                       "Congelante"        Moscu en invierno          ║
║   0 a 15                    "Frio"              Londres en primavera       ║
║   16 a 25                   "Templado"          CDMX todo el año           ║
║   26 a 35                   "Calido"            Miami en verano            ║
║   > 35                      "Extremo"           Dubai en agosto            ║
║                                                                            ║
╚════════════════════════════════════════════════════════════════════════════╝
```

**IMPORTANTE**: Los rangos son:
- `< 0`: Estrictamente menor que cero
- `0 a 15`: Desde 0 (inclusive) hasta 15 (inclusive) → `0 <= temp <= 15`
- `16 a 25`: Desde 16 (inclusive) hasta 25 (inclusive)
- `26 a 35`: Desde 26 (inclusive) hasta 35 (inclusive)
- `> 35`: Estrictamente mayor que 35

````python
def clasificar_temperatura(celsius):
    """Clasifica la temperatura segun rangos definidos."""
    if celsius < 0:
        return "Congelante"
    elif celsius <= 15:  # 0 a 15
        return "Frio"
    elif celsius <= 25:  # 16 a 25
        return "Templado"
    elif celsius <= 35:  # 26 a 35
        return "Calido"
    else:  # > 35
        return "Extremo"

# Pruebas de clasificacion
print("Tabla de clasificacion")
print("=" * 30)
temperaturas_prueba = [-15, -1, 0, 10, 15, 16, 22, 25, 26, 30, 35, 36, 45]
for temp in temperaturas_prueba:
    clasificacion = clasificar_temperatura(temp)
    print(f"{temp:>5}°C  ->  {clasificacion}")
````


### Regla 4: Formato de Temperatura en Salida

La temperatura en Celsius debe mostrarse con **1 decimal**.

```python
temperatura = 10.0
print(f"{temperatura:.1f}")  # Imprime: 10.0

temperatura = 22.333333

```

## Guia de Implementacion

### Paso 1: Leer el Encabezado

````python
import sys

# Leer y descartar la primera linea (encabezados de entrada)
# encabezado_entrada = input()  # o sys.stdin.readline()

# Imprimir encabezados de salida
print("ciudad,temperatura_celsius,clasificacion")
````

### Paso 2: Procesar Cada Linea

````python
# Ejemplo de como procesar una linea
linea = "CDMX,22,C"

# Separar por comas
partes = linea.strip().split(',')
print(f"Partes: {partes}")

# Verificar que hay 3 partes
if len(partes) == 3:
    ciudad = partes[0]
    temperatura_str = partes[1]
    unidad = partes[2].upper()  # Mayusculas para comparar
    
    print(f"Ciudad: {ciudad}")
    print(f"Temperatura: {temperatura_str}")
    print(f"Unidad: {unidad}")
````

### Paso 3: Validar y Convertir

````python
def procesar_linea(linea):
    """
    Procesa una linea del CSV.
    Retorna (ciudad, celsius, clasificacion) o None si es invalida.
    """
    # Separar
    partes = linea.strip().split(',')
    if len(partes) != 3:
        return None
    
    ciudad = partes[0]
    temp_str = partes[1]
    unidad = partes[2].strip().upper()
    
    # Validar unidad
    if unidad not in ['C', 'F']:
        return None
    
    # Convertir temperatura
    try:
        temperatura = float(temp_str)
    except ValueError:
        return None  # No es un numero valido
    
    # Convertir a Celsius si es necesario
    if unidad == 'F':
        celsius = (temperatura - 32) * 5 / 9
    else:
        celsius = temperatura
    
    # Clasificar
    if celsius < 0:
        clasificacion = "Congelante"
    elif celsius <= 15:
        clasificacion = "Frio"
    elif celsius <= 25:
        clasificacion = "Templado"
    elif celsius <= 35:
        clasificacion = "Calido"
    else:
        clasificacion = "Extremo"
    
    return (ciudad, celsius, clasificacion)

# Pruebas
lineas_prueba = [
    "CDMX,22,C",
    "Nueva York,50,F",
    "Error,abc,C",
    "Miami,95,X",
    "Moscu,-10,C"
]

for linea in lineas_prueba:
    resultado = procesar_linea(linea)
    if resultado:
        ciudad, celsius, clasif = resultado
        print(f"{ciudad},{celsius:.1f},{clasif}")
    else:
        print(f"[IGNORADA] {linea}")
````

### Paso 4: Programa Completo

````python
# ESTRUCTURA COMPLETA DEL PROGRAMA
# Guarda esto como main.py

codigo_ejemplo = '''
import sys

def fahrenheit_a_celsius(f):
    """Convierte Fahrenheit a Celsius."""
    return (f - 32) * 5 / 9

def clasificar(celsius):
    """Clasifica la temperatura."""
    if celsius < 0:
        return "Congelante"
    elif celsius <= 15:
        return "Frio"
    elif celsius <= 25:
        return "Templado"
    elif celsius <= 35:
        return "Calido"
    else:
        return "Extremo"

def main():
    # Leer y descartar encabezado de entrada
    primera_linea = True
    
    # Imprimir encabezado de salida
    print("ciudad,temperatura_celsius,clasificacion")
    
    for linea in sys.stdin:
        linea = linea.strip()
        
        # Saltar encabezado
        if primera_linea:
            primera_linea = False
            continue
        
        # Saltar lineas vacias
        if not linea:
            continue
        
        # Separar campos
        partes = linea.split(',')
        if len(partes) != 3:
            continue  # Ignorar linea invalida
        
        ciudad = partes[0]
        temp_str = partes[1]
        unidad = partes[2].strip().upper()
        
        # Validar unidad
        if unidad not in ['C', 'F']:
            continue  # Ignorar
        
        # Convertir temperatura
        try:
            temp = float(temp_str)
        except ValueError:
            continue  # Ignorar si no es numero
        
        # Convertir a Celsius
        if unidad == 'F':
            celsius = fahrenheit_a_celsius(temp)
        else:
            celsius = temp
        
        # Clasificar y imprimir
        clasificacion = clasificar(celsius)
        print(f"{ciudad},{celsius:.1f},{clasificacion}")

if __name__ == "__main__":
    main()
'''

print(codigo_ejemplo)
````

## Casos Borde a Considerar

Tu programa debe manejar correctamente:

2. **Temperaturas negativas**: -10, -0.5

## Errores Comunes

### 1. Olvidar el encabezado de salida
```python
# INCORRECTO - No imprime encabezado
for linea in sys.stdin:
    ...

# CORRECTO - Primero el encabezado
print("ciudad,temperatura_celsius,clasificacion")
for linea in sys.stdin:
    ...
```

### 2. No saltar el encabezado de entrada
```python
# INCORRECTO - Procesa "ciudad,temperatura,unidad" como dato
for linea in sys.stdin:
    procesar(linea)

# CORRECTO - Salta la primera linea
primera = True
for linea in sys.stdin:
    if primera:
        primera = False
        continue
    procesar(linea)
```

### 3. Confundir los rangos
```python
# INCORRECTO - 15 quedaria fuera!
if celsius < 15:
    return "Frio"

# CORRECTO - 15 esta incluido en Frio
if celsius <= 15:
    return "Frio"
```

### 4. No redondear a 1 decimal
```python
# INCORRECTO
print(f"{ciudad},{celsius},{clasificacion}")
# Salida: CDMX,22.333333333,Templado

# CORRECTO
print(f"{ciudad},{celsius:.1f},{clasificacion}")
# Salida: CDMX,22.3,Templado
```
