## El Patron Fundamental de Ciencia de Datos

Antes de empezar con el reto, entendamos algo importante. El 90% de los programas en ciencia de datos siguen este patron:

```
┌─────────────┐      ┌─────────────────┐      ┌─────────────┐
│   ENTRADA   │ ───> │   TU PROGRAMA   │ ───> │   SALIDA    │
│   (datos)   │      │   (procesa)     │      │ (resultado) │
└─────────────┘      └─────────────────┘      └─────────────┘
```

Esto se llama **ETL** (Extract, Transform, Load) o simplemente **pipeline de datos**:

1. **Leer** datos de alguna fuente (archivo, API, base de datos, stdin)
2. **Procesar** esos datos (limpiar, transformar, calcular)
3. **Escribir** el resultado (archivo, pantalla, base de datos, stdout)

Este reto, aunque sencillo, te enseñara este patron que usaras en TODOS los retos futuros.

## Reglas de Procesamiento

### Regla 1: Lineas Vacias
Si la linea esta vacia (o solo tiene espacios), imprimir `0`.

```
Entrada:  ""          (linea vacia)
Salida:   0

Entrada:  "   "       (solo espacios)
Salida:   0
```

### Regla 2: Un Solo Elemento
Si hay un solo numero, imprimirlo (truncado a entero).

```
Entrada:  "10"
Salida:   10

Entrada:  "3.9"
Salida:   3         (truncado, NO redondeado)
```

### Regla 3: Multiples Elementos
Sumar todos los valores y mostrar el resultado.

```
Entrada:  "1,2,3"
Salida:   6         (1 + 2 + 3)

Entrada:  "-5,10,3"
Salida:   8         (-5 + 10 + 3)
```

### Regla 4: Truncar Decimales (MUY IMPORTANTE)

Los numeros decimales deben **truncarse a entero ANTES de sumar**.

**Truncar** significa quitar los decimales, NO redondear:
- `3.9` -> `3` (no 4)
- `2.1` -> `2`
- `-2.7` -> `-2` (hacia cero, no hacia abajo)

```
Entrada:  "1.9,2.1,3.7"
Proceso:  1.9->1, 2.1->2, 3.7->3
Salida:   6         (1 + 2 + 3)

Entrada:  "-1.9,2.5"
Proceso:  -1.9->-1, 2.5->2
Salida:   1         (-1 + 2)
```

**En Python, usa `int()` para truncar:**

````python
# Diferencia entre truncar y redondear
print(f"int(3.9)   = {int(3.9)}")      # Truncar: 3
print(f"round(3.9) = {round(3.9)}")    # Redondear: 4

print(f"int(-2.7)  = {int(-2.7)}")     # Truncar: -2
print(f"round(-2.7)= {round(-2.7)}")   # Redondear: -3
````

### Regla 5: Limpiar Caracteres Invalidos

Si un valor contiene caracteres que NO son:
- Digitos (`0-9`)
- Punto decimal (`.`)
- Signo negativo (`-`)

Esos caracteres deben **eliminarse** antes de procesar.

```
Entrada:  "1a2,3b,4"
Limpieza: "12", "3", "4"
Salida:   19        (12 + 3 + 4)

Entrada:  "$100,USD50"
Limpieza: "100", "50"
Salida:   150       (100 + 50)

Entrada:  "abc"
Limpieza: ""        (no queda nada)
Proceso:  Se trata como 0
```

**Nota**: La coma es el separador, NO debe eliminarse.

### Regla 6: Ignorar Espacios

Los espacios antes y despues de cada valor deben ignorarse.

```
Entrada:  "  5 , 10 , 15  "
Limpieza: "5", "10", "15"
Salida:   30        (5 + 10 + 15)
```

**Usa `.strip()` en Python para quitar espacios.**

````python
# Filtrar caracteres (quedarse solo con digitos, punto y signo negativo)
def limpiar(texto):
    caracteres_validos = '0123456789.-'
    resultado = ''
    for char in texto:
        if char in caracteres_validos:
            resultado += char
    return resultado

# Prueba
print(limpiar('1a2b3'))      # '123'
print(limpiar('$100'))       # '100'
print(limpiar('-5.7abc'))    # '-5.7'
print(limpiar("aqwsd#$%3&/ -"))
````

````python
# Convertir string a numero y truncar
def convertir_a_entero(texto):
    """Convierte texto a entero, truncando decimales."""
    if not texto:  # Si esta vacio
        return 0
    try:
        numero = float(texto)  # Primero a float
        return int(numero)     # Luego truncar a int
    except ValueError:
        return 0  # Si no se puede convertir, es 0

# Pruebas
print(convertir_a_entero('10'))     # 10
print(convertir_a_entero('3.9'))    # 3
print(convertir_a_entero('-2.7'))   # -2
print(convertir_a_entero(''))       # 0
print(convertir_a_entero('3-'))       # 0
````

## Casos Borde a Considerar

Tu programa debe manejar correctamente:

1. **Linea completamente vacia** -> `0`
2. **Linea con solo espacios** -> `0`
3. **Valor que queda vacio despues de limpiar** (ej: `"abc"`) -> tratarlo como `0`
4. **Multiples comas seguidas** (ej: `"1,,2"`) -> el valor vacio es `0`
5. **Coma al inicio o final** (ej: `",1,2"` o `"1,2,"`) -> valores vacios son `0`
6. **Numeros muy grandes** -> Python maneja enteros grandes automaticamente
7. **Punto decimal sin parte entera** (ej: `".5"`) -> `0` (truncar 0.5)
8. **Multiples puntos** (ej: `"1.2.3"`) -> tu programa decidira como manejarlo

## Tips Finales

### Desarrollo Iterativo

1. **Empieza simple**: Haz que funcione para el caso basico (`1,2,3`)
2. **Agrega complejidad**: Maneja decimales, luego caracteres basura
3. **Prueba constantemente**: Despues de cada cambio, ejecuta pruebas
4. **Haz commits frecuentes**: Guarda tu progreso en Git

### Errores Comunes

- **Olvidar `strip()`**: Las lineas de stdin traen `\n` al final
- **Usar `round()` en lugar de `int()`**: Debes truncar, no redondear
- **Imprimir mensajes extra**: Solo imprime el numero, nada mas
- **No manejar lineas vacias**: Una linea vacia debe dar `0`
