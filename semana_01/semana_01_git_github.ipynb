## Que es Git?

Git es un **sistema de control de versiones distribuido**. Vamos a desglosar eso:

- **Control de versiones**: Guarda "fotos" (snapshots) de tu proyecto en diferentes momentos del tiempo
- **Distribuido**: Cada persona tiene una copia completa del historial, no solo los archivos actuales

### Analogia: Git como un Album de Fotos

Imagina que tu proyecto es como tu vida y Git es un album fotografico:

- Cada **commit** es como tomar una foto de un momento importante
- El **repositorio** es el album completo con todas las fotos ordenadas cronologicamente
- Las **ramas (branches)** son como lineas de tiempo alternativas ("que hubiera pasado si...")
- Puedes **volver atras** a cualquier foto en cualquier momento

### Por que Git es ESENCIAL en Ciencia de Datos?

1. **Reproducibilidad**: Puedes volver a cualquier version de tu analisis
2. **Colaboracion**: Varios cientificos de datos pueden trabajar en el mismo proyecto
3. **Experimentacion segura**: Prueba ideas locas sin miedo a romper lo que funciona
4. **Documentacion automatica**: El historial cuenta la historia de tu proyecto
5. **Estandar de la industria**: El 95%+ de empresas tech lo usan

## Los 3 Estados de Git: Donde Viven tus Archivos

Este es el concepto MAS IMPORTANTE para entender Git. Tus archivos pueden estar en 3 lugares diferentes:

### 1. Working Directory (Directorio de Trabajo)

Es simplemente la carpeta de tu proyecto en tu computadora. Aqui es donde:
- Creas archivos nuevos
- Editas codigo existente
- Eliminas lo que no necesitas

**Los archivos aqui pueden estar:**
- **Untracked** (sin seguimiento): Git no sabe que existen
- **Modified** (modificados): Git los conoce, pero detecta cambios no guardados
- **Unmodified**: Identicos a la ultima version en el repositorio

### 2. Staging Area (Area de Preparacion)

Piensa en esto como una **sala de espera** o un **carrito de compras**:
- Aqui pones los archivos que quieres incluir en tu proximo commit
- Puedes agregar algunos archivos y dejar otros fuera
- Te da control sobre QUE entra en cada "foto"

**Por que existe?**
- Permite hacer commits parciales (no todo o nada)
- Puedes revisar que vas a guardar antes de hacerlo
- Separa "editar" de "guardar definitivamente"

### 3. Repository (Repositorio - carpeta .git)

Es la **base de datos** donde Git guarda todo el historial:
- Cada commit con sus metadatos (autor, fecha, mensaje)
- Todas las versiones de todos los archivos
- La estructura de ramas

**Vive en una carpeta oculta llamada `.git`** - NUNCA la modifiques manualmente!

## Comandos Esenciales de Git (El 80/20)

Estos son los comandos que usaras el 80% del tiempo. Dominandolos, podras trabajar profesionalmente.

### Configuracion Inicial (solo una vez)

```bash
# Decirle a Git quien eres (aparece en cada commit)
git config --global user.name "Tu Nombre"
git config --global user.email "tu.email@ejemplo.com"

# Ver tu configuracion actual
git config --list
```

### Crear y Clonar Repositorios

```bash
# Crear un nuevo repositorio en la carpeta actual
git init

# Clonar un repositorio existente de internet
git clone https://github.com/usuario/repositorio.git
```

**`git init`** crea la carpeta oculta `.git` y convierte tu carpeta en un repositorio.

**`git clone`** descarga una copia completa de un repositorio remoto, incluyendo todo el historial.

### Ver Estado y Cambios

```bash
# Ver el estado actual (MUY IMPORTANTE - usalo constantemente!)
git status

# Ver cambios en archivos modificados (que NO estan en staging)
git diff

# Ver cambios en archivos que YA estan en staging
git diff --staged

# Ver historial de commits
git log

# Ver historial compacto (una linea por commit)
git log --oneline

# Ver historial con grafico de ramas
git log --oneline --graph --all
```

**Pro tip:** `git status` es tu mejor amigo. Ejecutalo ANTES y DESPUES de cada operacion para entender que esta pasando.

### Agregar Archivos al Staging

```bash
# Agregar un archivo especifico
git add archivo.py

# Agregar multiples archivos
git add archivo1.py archivo2.py carpeta/

# Agregar todos los archivos modificados y nuevos
git add .

# Agregar solo archivos ya rastreados (no nuevos)
git add -u
```

**Advertencia sobre `git add .`**: Agrega TODO, incluyendo archivos que quizas no quieres (como contraseñas o archivos temporales). Usa `.gitignore` para protegerte (lo veremos mas adelante).

### Crear Commits

```bash
# Crear un commit con mensaje
git commit -m "Agrega funcion para calcular promedio"

# Agregar al staging Y hacer commit en un solo paso (solo archivos ya rastreados)
git commit -am "Corrige error en validacion de datos"
```

### Buenos Mensajes de Commit

```
   MALO                              BUENO
   ─────────────────────────────────────────────────────────
   "cambios"                    ->  "Agrega validacion de email"
   "fix"                        ->  "Corrige error de division por cero"
   "asdfasdf"                   ->  "Actualiza dependencias de pandas"
   "WIP"                        ->  "Implementa lectura de CSV"
```

**Estructura recomendada:** `Verbo + Que` (en presente, imperativo)
- "Agrega...", "Corrige...", "Elimina...", "Actualiza...", "Refactoriza..."

### Deshacer Cambios

```bash
# Quitar un archivo del staging (sin perder cambios)
git restore --staged archivo.py

# Descartar cambios en un archivo (PELIGRO: pierdes los cambios!)
git restore archivo.py

# Ver un archivo como estaba en un commit anterior
git show HEAD~1:archivo.py
```

**Nota de seguridad:** Una vez que haces commit, tus cambios estan a salvo. Los comandos peligrosos son los que afectan cambios NO commiteados.

### Trabajar con Ramas (Branches)

Las ramas te permiten trabajar en diferentes versiones de tu proyecto simultaneamente.

```bash
# Ver ramas existentes (* indica la actual)
git branch

# Crear una nueva rama
git branch nombre-de-rama

# Cambiar a otra rama
git checkout nombre-de-rama

# Crear Y cambiar a nueva rama (atajo)
git checkout -b nombre-de-rama

# Unir otra rama a la actual
git merge nombre-de-rama

# Eliminar una rama (ya fusionada)
git branch -d nombre-de-rama
```

```
              main                 feature-nueva
                │                        │
                ▼                        ▼
    ●───●───●───●                        
                 \                       
                  ●───●───●              <- Trabajas aqui sin afectar main
                           \             
    ●───●───●───●───────────●            <- Merge: juntas los cambios
```

### Sincronizar con Repositorio Remoto (GitHub)

```bash
# Ver repositorios remotos configurados
git remote -v

# Agregar un repositorio remoto
git remote add origin https://github.com/usuario/repo.git

# Descargar cambios del remoto (sin fusionar)
git fetch

# Descargar Y fusionar cambios del remoto
git pull

# Subir tus commits al remoto
git push

# Primera vez que subes una rama nueva
git push -u origin nombre-de-rama
```

## El Archivo .gitignore: Tu Guardian de Secretos

El archivo `.gitignore` le dice a Git que archivos debe IGNORAR completamente. Esto es crucial para:

1. **Seguridad**: No subir contraseñas, API keys, archivos .env
2. **Limpieza**: No incluir archivos generados, cache, dependencias
3. **Eficiencia**: No trackear archivos grandes o binarios innecesarios

### Ejemplo de .gitignore para Ciencia de Datos

```gitignore
# Archivos de entorno y secretos
.env
config.ini
secrets/
*.pem

# Ambiente virtual de Python
.venv/
venv/
env/

# Cache de Python
__pycache__/
*.py[cod]
*.pyo
.pytest_cache/

# Jupyter Notebook
.ipynb_checkpoints/

# Archivos de IDE
.vscode/
.idea/
*.swp

# Datos grandes (descargados, no originales)
data/raw/*.csv
data/external/
*.parquet

# Modelos entrenados (muy grandes)
models/*.pkl
models/*.joblib

# Outputs regenerables
outputs/
*.log

# Sistema operativo
.DS_Store
Thumbs.db
```

### Patrones Comunes en .gitignore

```
*.log           # Cualquier archivo que termine en .log
temp/           # Carpeta completa llamada temp
!importante.log # EXCEPCION: si incluir este archivo
data/*.csv      # Todos los .csv dentro de data/
**/cache/       # Cualquier carpeta llamada cache, en cualquier nivel
```

## Errores Comunes y Como Solucionarlos

### "No me deja hacer push"

**Problema**: `error: failed to push some refs`

**Solucion**: Alguien mas (o tu desde otro lugar) subio cambios. Primero baja los cambios:
```bash
git pull
# Resuelve conflictos si hay
git push
```

### "Agregue un archivo que no debia"

**Solucion**: Quitalo del staging (sin borrar el archivo):
```bash
git restore --staged archivo_secreto.txt
```

### "Hice commit de algo incorrecto"

**Solucion** (si NO has hecho push):
```bash
git reset --soft HEAD~1  # Deshace el commit, mantiene cambios en staging
```

### "El .gitignore no ignora un archivo"

**Problema**: El archivo ya estaba siendo rastreado antes de agregarlo al .gitignore

**Solucion**:
```bash
git rm --cached archivo.txt  # Lo quita del tracking (sin borrar)
git commit -m "Deja de rastrear archivo.txt"
```
