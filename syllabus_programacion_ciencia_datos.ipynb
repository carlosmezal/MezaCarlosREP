### Reto Semana 1: Calculadora de Sumas

**Contexto**: Tu primer programa procesara un archivo de entrada linea por linea. Este patron (leer entrada -> procesar -> escribir salida) es fundamental en ciencia de datos.

#### Entrada
- Se lee desde **entrada estandar (stdin)**, linea por linea hasta fin de archivo (EOF)
- Cada linea contiene una lista de valores separados por comas
- Puede haber entre 0 y 1000 elementos por linea

#### Reglas de procesamiento
1. **Lineas vacias**: Imprimir `0`
2. **Un solo elemento**: Imprimir ese numero (truncado a entero)
3. **Multiples elementos**: Sumar todos y imprimir el resultado
4. **Numeros flotantes**: Truncar a entero ANTES de sumar (no redondear)
   - Ejemplo: `3.9` se convierte en `3`
5. **Caracteres invalidos**: Eliminar cualquier caracter que no sea digito, punto, coma o signo negativo
   - Ejemplo: `1a2,3b4` -> `12,34` -> `12 + 34 = 46`
6. **Espacios**: Ignorar espacios antes/despues de numeros

#### Salida
- Imprimir en **salida estandar (stdout)**, una linea por cada linea de entrada
- Solo el numero resultante, sin texto adicional

#### Ejemplos

| Entrada | Salida | Explicacion |
|---------|--------|-------------|
| `1,2,3` | `6` | 1+2+3=6 |
| `10` | `10` | Un solo elemento |
| `` (vacia) | `0` | Linea vacia |
| `1.9,2.1,3.7` | `6` | 1+2+3=6 (truncados) |
| `1a2,3b,4` | `16` | 12+3+4=19... wait, let me recalculate |
| `-5,10,3` | `8` | -5+10+3=8 |
| `  5 , 10 , 15  ` | `30` | Espacios ignorados |

#### Ejecucion
```bash
# Desde archivo
python main.py < entrada.txt

# O con pipe
cat entrada.txt | python main.py
```

#### Entregable
- Repositorio en GitHub con minimo 3 commits significativos
- Archivo `main.py` que lea de stdin y escriba a stdout
- `README.md` con instrucciones de uso

### Reto Semana 2: Clasificador de Temperaturas

**Contexto**: Tienes un archivo con mediciones de temperatura de diferentes ciudades. Debes clasificar cada medicion y generar un reporte.

#### Entrada
- Archivo CSV desde stdin
- Columnas: `ciudad,temperatura,unidad`
- `unidad` puede ser `C` (Celsius) o `F` (Fahrenheit)

#### Reglas de procesamiento
1. Convertir todas las temperaturas a Celsius
   - Formula: `C = (F - 32) * 5/9`
2. Clasificar segun temperatura en Celsius:
   - `< 0`: "Congelante"
   - `0-15`: "Frio"
   - `16-25`: "Templado"
   - `26-35`: "Calido"
   - `> 35`: "Extremo"
3. Ignorar lineas con datos invalidos (temperatura no numerica)

#### Salida
- CSV a stdout
- Columnas: `ciudad,temperatura_celsius,clasificacion`
- Temperatura redondeada a 1 decimal

#### Ejemplo
**Entrada:**
```
ciudad,temperatura,unidad
CDMX,22,C
Nueva York,50,F
Moscu,-10,C
Miami,95,F
```

**Salida:**
```
ciudad,temperatura_celsius,clasificacion
CDMX,22.0,Templado
Nueva York,10.0,Frio
Moscu,-10.0,Congelante
Miami,35.0,Calido
```

## Semana 6: Expresiones Regulares - Fundamentos

### Objetivo
Comprender la sintaxis de expresiones regulares y aplicarlas para validar y extraer patrones simples en texto.

### Descripcion
Las expresiones regulares (regex) son una herramienta poderosa para trabajar con texto. En ciencia de datos las usamos constantemente para limpiar datos, validar formatos y extraer informacion estructurada de texto no estructurado.

### Temas a desarrollar

1. **Introduccion a regex**
   - Que es un patron
   - Modulo `re` de Python
   - Raw strings: `r"patron"`

2. **Metacaracteres basicos**
   - `.`: Cualquier caracter
   - `^`: Inicio de linea
   - `$`: Fin de linea
   - `\d`: Digito (0-9)
   - `\w`: Alfanumerico (a-z, A-Z, 0-9, _)
   - `\s`: Espacio en blanco

3. **Cuantificadores**
   - `*`: 0 o mas
   - `+`: 1 o mas
   - `?`: 0 o 1
   - `{n}`: Exactamente n
   - `{n,m}`: Entre n y m

4. **Clases de caracteres**
   - `[abc]`: a, b o c
   - `[a-z]`: Rango
   - `[^abc]`: Negacion

5. **Funciones de `re`**
   - `re.match()`: Coincidencia al inicio
   - `re.search()`: Primera coincidencia
   - `re.findall()`: Todas las coincidencias
   - `re.sub()`: Reemplazar

6. **Patrones comunes en datos**
   - Telefonos: `\d{10}`
   - Codigos postales: `\d{5}`
   - Fechas simples: `\d{4}-\d{2}-\d{2}`

## Semana 7: Expresiones Regulares - Avanzado

### Objetivo
Aplicar expresiones regulares avanzadas para limpiar y normalizar datos del mundo real: emails, telefonos, direcciones, identificadores.

### Descripcion
Los datos reales vienen en formatos inconsistentes. Esta semana aprenderemos patrones avanzados para manejar la variabilidad del mundo real y normalizar informacion de contacto.

### Temas a desarrollar

1. **Grupos de captura**
   - Parentesis para agrupar: `(patron)`
   - Acceso a grupos: `match.group(1)`
   - Grupos nombrados: `(?P<nombre>patron)`

2. **Patrones avanzados**
   - Alternancia: `patron1|patron2`
   - Grupos no capturadores: `(?:patron)`
   - Lookahead: `(?=patron)`, `(?!patron)`
   - Lookbehind: `(?<=patron)`, `(?<!patron)`

3. **Validacion de emails**
   ```python
   r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
   ```
   - Partes: usuario, dominio, extension
   - Casos borde comunes

4. **Normalizacion de telefonos**
   - Eliminar caracteres no numericos
   - Manejar codigos de pais (52, +52, 521)
   - Extraer LADA y numero
   - Formatear consistentemente

5. **Limpieza de texto**
   - Eliminar caracteres especiales
   - Normalizar espacios multiples
   - Capitalizar nombres propios
   - Estandarizar formatos de fecha

6. **Flags de regex**
   - `re.IGNORECASE`: Ignorar mayusculas/minusculas
   - `re.MULTILINE`: `^` y `$` por linea
   - `re.DOTALL`: `.` incluye saltos de linea

## Semana 9: NumPy y Estadistica Descriptiva

### Objetivo
Aplicar NumPy para calculos estadisticos avanzados, detectar outliers y analizar distribuciones de datos.

### Descripcion
La estadistica descriptiva nos permite entender nuestros datos antes de cualquier analisis. Esta semana profundizaremos en metricas avanzadas y tecnicas para identificar valores atipicos.

### Temas a desarrollar

1. **Percentiles y cuartiles**
   - `np.percentile()`, `np.quantile()`
   - Q1 (25%), Q2/mediana (50%), Q3 (75%)
   - Interpretacion de percentiles

2. **Rango intercuartilico (IQR)**
   - `IQR = Q3 - Q1`
   - Limite inferior: `Q1 - 1.5 * IQR`
   - Limite superior: `Q3 + 1.5 * IQR`
   - Deteccion de outliers

3. **Distribucion de datos**
   - `np.histogram()`: frecuencias
   - Asimetria (skewness)
   - Curtosis
   - Normalidad

4. **Correlacion**
   - `np.corrcoef()`: matriz de correlacion
   - Interpretacion: -1 a 1
   - `np.cov()`: covarianza

5. **Operaciones matriciales**
   - `np.dot()`: producto punto
   - `np.transpose()` o `.T`
   - Operaciones sobre ejes: `axis=0`, `axis=1`

6. **Generacion de datos**
   - `np.random.seed()`: reproducibilidad
   - `np.random.normal()`, `np.random.uniform()`
   - Muestreo: `np.random.choice()`
