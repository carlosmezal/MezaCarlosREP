## 1. Ambientes Virtuales con venv

Un **ambiente virtual** es una carpeta aislada que contiene:
- Una copia del interprete de Python
- Sus propias librerias instaladas
- Independiente del sistema y otros proyectos

### Crear un Ambiente Virtual

````python
# Comandos para la terminal (no ejecutar aqui)

comandos_venv = '''
# 1. CREAR ambiente virtual (en la carpeta del proyecto)
#    Esto crea una carpeta .venv con Python aislado

python -m venv .venv

# Estructura creada:
# .venv/
# ├── Include/
# ├── Lib/
# │   └── site-packages/    <- Aqui se instalan las librerias
# ├── Scripts/              <- Windows
# │   ├── activate.bat
# │   ├── python.exe
# │   └── pip.exe
# └── pyvenv.cfg
'''

print(comandos_venv)
````

````python
# 2. ACTIVAR el ambiente virtual

activar = '''
# Windows (PowerShell)
.venv\\Scripts\\Activate.ps1

# Windows (CMD)
.venv\\Scripts\\activate.bat

# Linux / Mac
source .venv/bin/activate

# Sabras que esta activado porque el prompt cambia:
# (.venv) C:\\mi_proyecto>
#  ^^^^^^ Esto indica el ambiente activo
'''

print(activar)
````

````python
# 3. DESACTIVAR el ambiente

desactivar = '''
# Simplemente escribe:
deactivate

# El prompt vuelve a la normalidad:
# C:\\mi_proyecto>
'''

print(desactivar)
````

````python
# 4. VERIFICAR que estas en el ambiente correcto

verificar = '''
# Ver que Python estas usando
where python      # Windows
which python      # Linux/Mac

# Deberia mostrar algo como:
# C:\\mi_proyecto\\.venv\\Scripts\\python.exe

# Ver que pip estas usando
where pip         # Windows
which pip         # Linux/Mac

# Deberia mostrar:
# C:\\mi_proyecto\\.venv\\Scripts\\pip.exe
'''

print(verificar)
````

### Errores Comunes con PowerShell

## 5. Implementando un Perfilador Basico (Sin Librerias)

Vamos a construir un perfilador usando solo Python puro.

````python
# Datos de ejemplo para perfilar

datos_ejemplo = [
    ["id", "nombre", "edad", "email", "salario", "fecha_ingreso"],
    ["1", "Ana Garcia", "28", "ana@email.com", "35000.50", "2023-01-15"],
    ["2", "Bob Lopez", "", "bob@email.com", "42000", "2022-06-01"],
    ["3", "Carlos Ruiz", "35", "", "38500.75", "2023-03-20"],
    ["4", "Diana Mora", "29", "diana@email.com", "", "2021-11-10"],
    ["5", "Eva Torres", "31", "eva@email.com", "45000", ""],
    ["6", "Ana Garcia", "28", "ana2@email.com", "35000.50", "2024-01-01"],  # Nombre duplicado
    ["7", "Frank Villa", "abc", "frank@email.com", "40000", "2023-08-15"],  # Edad invalida
    ["8", "", "25", "test@email.com", "32000", "2023-05-01"],  # Nombre vacio
]

# Mostrar datos
print("Datos de ejemplo:")
for fila in datos_ejemplo:
    print(fila)
````

````python
def es_vacio(valor):
    """Determina si un valor se considera vacio/nulo."""
    if valor is None:
        return True
    if isinstance(valor, str) and valor.strip() == "":
        return True
    return False


def inferir_tipo(valores):
    """
    Infiere el tipo de dato de una columna.
    Retorna: 'numerico', 'texto', 'fecha', 'booleano'
    """
    # Filtrar valores no vacios
    valores_validos = [v for v in valores if not es_vacio(v)]
    
    if not valores_validos:
        return "vacio"
    
    # Contar tipos
    num_numericos = 0
    num_fechas = 0
    num_booleanos = 0
    
    for v in valores_validos:
        v_str = str(v).strip().lower()
        
        # Verificar si es numerico
        try:
            float(v_str.replace(',', ''))
            num_numericos += 1
            continue
        except ValueError:
            pass
        
        # Verificar si es fecha (formato YYYY-MM-DD)
        if len(v_str) == 10 and v_str[4] == '-' and v_str[7] == '-':
            try:
                partes = v_str.split('-')
                if len(partes) == 3:
                    año, mes, dia = int(partes[0]), int(partes[1]), int(partes[2])
                    if 1900 <= año <= 2100 and 1 <= mes <= 12 and 1 <= dia <= 31:
                        num_fechas += 1
                        continue
            except ValueError:
                pass
        
        # Verificar si es booleano
        if v_str in ['true', 'false', 'yes', 'no', 'si', '1', '0']:
            num_booleanos += 1
    
    total = len(valores_validos)
    
    # Decidir tipo basado en mayoria (>80%)
    if num_numericos / total > 0.8:
        return "numerico"
    elif num_fechas / total > 0.8:
        return "fecha"
    elif num_booleanos / total > 0.8:
        return "booleano"
    else:
        return "texto"


# Probar inferencia de tipo
print("Pruebas de inferencia de tipo:")
print(f"['1', '2', '3']: {inferir_tipo(['1', '2', '3'])}")
print(f"['Ana', 'Bob']: {inferir_tipo(['Ana', 'Bob'])}")
print(f"['2023-01-01', '2023-02-15']: {inferir_tipo(['2023-01-01', '2023-02-15'])}")
print(f"['', '', '']: {inferir_tipo(['', '', ''])}")
````

````python
def perfilar_columna(nombre_columna, valores):
    """
    Genera el perfil de una columna.
    
    Args:
        nombre_columna: Nombre de la columna
        valores: Lista de valores de la columna
    
    Returns:
        dict: Diccionario con las metricas del perfil
    """
    total = len(valores)
    
    # Contar nulos
    nulos = sum(1 for v in valores if es_vacio(v))
    
    # Valores no nulos
    valores_validos = [v for v in valores if not es_vacio(v)]
    
    # Valores unicos
    unicos = len(set(valores_validos))
    
    # Primer valor no nulo como ejemplo
    ejemplo = valores_validos[0] if valores_validos else ""
    
    # Tipo inferido
    tipo = inferir_tipo(valores)
    
    return {
        "nombre_columna": nombre_columna,
        "tipo_inferido": tipo,
        "total_registros": total,
        "valores_nulos": nulos,
        "porcentaje_nulos": round(nulos / total * 100, 2) if total > 0 else 0,
        "valores_unicos": unicos,
        "porcentaje_unicos": round(unicos / total * 100, 2) if total > 0 else 0,
        "ejemplo_valor": ejemplo
    }


def perfilar_dataset(datos):
    """
    Perfila un dataset completo.
    
    Args:
        datos: Lista de listas donde la primera fila son encabezados
    
    Returns:
        list: Lista de perfiles (uno por columna)
    """
    if not datos or len(datos) < 2:
        return []
    
    encabezados = datos[0]
    filas = datos[1:]
    
    perfiles = []
    
    # Para cada columna
    for i, nombre_col in enumerate(encabezados):
        # Extraer valores de esta columna
        valores = [fila[i] if i < len(fila) else "" for fila in filas]
        
        # Generar perfil
        perfil = perfilar_columna(nombre_col, valores)
        perfiles.append(perfil)
    
    return perfiles
````

````python
# Ejecutar perfilado
perfiles = perfilar_dataset(datos_ejemplo)

# Mostrar resultados
print("=" * 80)
print("PERFIL DEL DATASET")
print("=" * 80)

for p in perfiles:
    print(f"\n--- {p['nombre_columna']} ---")
    print(f"  Tipo inferido:      {p['tipo_inferido']}")
    print(f"  Total registros:    {p['total_registros']}")
    print(f"  Valores nulos:      {p['valores_nulos']} ({p['porcentaje_nulos']}%)")
    print(f"  Valores unicos:     {p['valores_unicos']} ({p['porcentaje_unicos']}%)")
    print(f"  Ejemplo:            '{p['ejemplo_valor']}'")
````

````python
# Generar salida CSV

print("\n" + "=" * 80)
print("SALIDA CSV:")
print("=" * 80)

# Encabezados
campos = ["nombre_columna", "tipo_inferido", "total_registros", 
          "valores_nulos", "porcentaje_nulos", "valores_unicos",
          "porcentaje_unicos", "ejemplo_valor"]

print(",".join(campos))

# Datos
for p in perfiles:
    valores = [str(p[campo]) for campo in campos]
    print(",".join(valores))
````
